```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('country_wise_latest.csv')
```

```python
df.columns = df.columns.str.strip()
df['Country/Region'] = df['Country/Region'].str.strip().str.title()
df['WHO Region'] = df['WHO Region'].str.strip().str.title()

print(df.head())
```

```python
sns.histplot(df['Confirmed'], bins=30, kde=True)
plt.title('Distribution of Confirmed Cases')
plt.show()
```

```python
sns.boxplot(y=df['Deaths'])
plt.title('Boxplot of Deaths')
plt.show()
```

```python
top10 = df.sort_values('Confirmed', ascending=False).head(10)
sns.barplot(x='Confirmed', y='Country/Region', data=top10)
plt.title('Top 10 Countries by Confirmed Cases')
plt.show()
```

```python
top5 = df.sort_values('Confirmed', ascending=False).head(5)
plt.pie(top5['Confirmed'], labels=top5['Country/Region'], autopct='%1.1f%%')
plt.title('Top 5 Countries Share')
plt.show()
```

```python
region_confirmed = df.groupby('WHO Region')['Confirmed'].sum().reset_index()
sns.barplot(x='Confirmed', y='WHO Region', data=region_confirmed)
plt.title('Confirmed Cases by WHO Region')
plt.show()
```

```python
region_deaths = df.groupby('WHO Region')['Deaths'].sum().reset_index()
sns.barplot(x='Deaths', y='WHO Region', data=region_deaths)
plt.title('Deaths by WHO Region')
plt.show()
```

```python
top5.set_index('Country/Region')[['Confirmed','Deaths','Recovered']].plot(kind='bar', stacked=True)
plt.title('Confirmed, Deaths, Recovered - Top 5')
plt.show()
```

```python
sns.scatterplot(x='Confirmed', y='Deaths', data=df)
plt.title('Confirmed vs Deaths')
plt.show()
```

```python
sns.lineplot(x='Confirmed', y='Recovered', data=df)
plt.title('Confirmed vs Recovered')
plt.show()
```

```python
sns.heatmap(df[['Confirmed','Deaths','Recovered','Active']].corr(), annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.pairplot(df[['Confirmed','Deaths','Recovered','Active']])
plt.show()
```

```python
sns.boxplot(x='WHO Region', y='Confirmed', data=df)
plt.xticks(rotation=45)
plt.title('Confirmed Cases by WHO Region')
plt.show()
```

```python
sns.violinplot(x='WHO Region', y='Deaths', data=df)
plt.xticks(rotation=45)
plt.title('Deaths Distribution by Region')
plt.show()
```

```python
sns.histplot(df['Active'], bins=30, kde=True)
plt.title('Distribution of Active Cases')
plt.show()
```

```python
sns.lineplot(x='Country/Region', y='Confirmed',data=top10.sort_values('Confirmed'))
plt.xticks(rotation=45)
plt.title('Confirmed Cases - Top 10 Countries')
plt.show()
```
